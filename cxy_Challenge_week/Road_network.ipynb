```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# 1. 读取路网 GeoJSON 文件
geojson_path = "Road_network.geojson"
roads = gpd.read_file(geojson_path)

# 2. 打印数据基本信息
print(roads.head())
print("坐标系:", roads.crs)

# 3. 绘制路网
fig, ax = plt.subplots(figsize=(10, 10))
roads.plot(ax=ax, linewidth=0.5, color="black")
ax.set_title("Road Network", fontsize=16)
ax.set_axis_off()
plt.show()
```
